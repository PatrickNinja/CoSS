```python
import os
import re
import pandas as pd
from tqdm.auto import tqdm
```

```python
def train_findfiles(path):
    # 首先遍历当前目录所有文件及文件夹
    file_list = os.listdir(path)
    # 循环判断每个元素是否是文件夹还是文件，是文件夹的话，递归
    for file in file_list:
        # 利用os.path.join()方法取得路径全名，并存入cur_path变量，否则每次只能遍历一层目录
        cur_path = os.path.join(path, file)
        # 判断是否是文件夹
        if os.path.isdir(cur_path):
            train_findfiles(cur_path)
        else:
            # 判断是否是特定文件名称
            if ".dot" in file:
                filename = int(file.split("-")[0].replace('Java', ""))
                node_list = []
                edge_list = []
                with open(cur_path, "r") as f:
                    code = f.read()
                    code = code.split("\n")
                    for text in code:
                        text = text.strip().replace('[label="False"];', "").replace('[label="True"];', "")
                        if text.startswith("v"):
                            if "label" in text:
                                text1 = text.split(" ")
                                n = text1[0]
                                text = text.replace(n, "").replace('[label="', "").replace('"];', "")
                                text2 = text.strip()
                                text3 = text2.split(" ")
                                if ":" in text3[0]:
                                    text2 = text2.replace(text3[0], "").strip()
                                node_list.append(text2)
                            elif "->" in text:
                                text1 = text.replace(';', "")
                                text2 = text1.split(" ")
                                n1, n2 = text2[0], text2[2]
                                n1 = int(n1.replace('v', ""))
                                n2 = int(n2.replace('v', ""))
                                n1, n2 = n1 - 1, n2 - 1
                                edge_list.append([n1, n2])
                    train_D.append([filename, node_list, edge_list])
```
